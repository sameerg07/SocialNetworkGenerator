```python
from datetime import datetime
import pandas as pd
f = open("wiki-Rfa.txt",encoding="utf8")

sources = []
targets = []
votes = []
results = []
years = []
dates = []
texts = []
for line in f:
    if(line[0:3] == "SRC"):
        sources.append(line[4:])
    if(line[0:3] == "TGT"):
        targets.append(line[4:])
    if(line[0:3] == "VOT"):
        votes.append(line[4:])
    if(line[0:3] == "RES"):
        results.append(line[4:])
    if(line[0:3] == "YEA"):
        years.append(line[4:])
    if(line[0:3] == "DAT"):
        dates.append(line[4:])
    if(line[0:3] == "TXT"):
        texts.append(line[4:])
        
df = pd.DataFrame()
df['src'] = sources
df['tgt'] = targets
df['vote'] = votes
df['res'] = results
df['year'] = years
df['date'] = dates
df['txt'] = texts
```

```python
print(df)
```

```python
nodes_train = list(set(df['src']).union(set(df['tgt'])))
len(nodes_train)

edges_train = [[df.iloc[i]['src'],df.iloc[i]['tgt'],df.iloc[i]['vote']] for i in range(len(df))] 

# Considering neutral vote as -ve edge

fp1 = open("wiki_Rfa_full.gml",'w',encoding="utf8")
fp1.write("graph\n[\n  multigraph 1\n")
for i in nodes_train:
    fp1.write("  node\n  [\n    id "+i+"\n  ]\n")
    
for i in edges_train:
    fp1.write("  edge\n  [\n    source "+i[0])
    fp1.write("\n    target "+i[1])
    if int(i[2]) in [-1,0]:
        fp1.write("\n    weight 0\n  ]\n")
    else:
        fp1.write("\n    weight 1\n  ]\n")
fp1.write("\n]") 
fp1.close()
```
